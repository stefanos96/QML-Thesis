```python
#ΚΩΔΙΚΑΣ 4.5.1
import numpy as np

# ΑΠΛΟ ΠΑΡΑΔΕΙΓΜΑ ΚΒΑΝΤΙΚΟΥ ΜΟΝΤΕΛΟΥ
# Η έξοδος παριστάνει μια αναμενόμενη τιμή, όπως συμβαίνει σε QML μοντέλα.
def model(x, params):
    return np.cos(x + np.sum(params))

# ΣΥΝΑΡΤΗΣΗ ΜΕΤΑΤΟΠΙΣΗΣ ΠΑΡΑΜΕΤΡΟΥ
def parameter_shift_derivative(model, x, params, index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shift = np.pi / 2

    shifted_plus[index] += shift
    shifted_minus[index] -= shift

    y_plus = model(x, shifted_plus)
    y_minus = model(x, shifted_minus)

    derivative = 0.5 * (y_plus - y_minus)

    return derivative

# ΥΠΟΛΟΓΙΣΜΟΣ GRADIENT ΓΙΑ ΟΛΕΣ ΤΙΣ ΠΑΡΑΜΕΤΡΟΥΣ
def parameter_shift_gradient(model, x, params):
    gradients = np.zeros_like(params)

    for index in range(len(params)):
        gradients[index] = parameter_shift_derivative(model, x, params, index)

    return gradients

# ΣΥΝΑΡΤΗΣΗ ΕΝΗΜΕΡΩΣΗΣ ΠΑΡΑΜΕΤΡΩΝ ΑΠΟ ΤΟ 4.4
def gradient_descent_update(params, gradients, learning_rate):
    params = np.array(params)
    gradients = np.array(gradients)

    updated_params = params - learning_rate * gradients
    return updated_params

# ΠΑΡΑΔΕΙΓΜΑ ΕΚΤΕΛΕΣΗΣ
x = 0.5
params = np.array([0.3, 1.2])
learning_rate = 0.1

print("Αρχικές παράμετροι:", params)

y_pred = model(x, params)
print("Έξοδος μοντέλου:", y_pred)

gradients = parameter_shift_gradient(model, x, params)
print("Gradient:", gradients)

updated_params = gradient_descent_update(params, gradients, learning_rate)
print("Νέες παράμετροι:", updated_params)
```

```python
# ΚΩΔΙΚΑΣ 4.6 ΓΙΑ 2 QUBITS ΜΕ 2 VARIATIONAL LAYERS

import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp


# ΣΤΑΘΕΡΕΣ ΜΟΝΤΕΛΟΥ
N_QUBITS = 2
N_LAYERS = 2


# ΚΒΑΝΤΙΚΟ ΜΟΝΤΕΛΟ ΠΟΛΛΩΝ QUBITS ΜΕ ΠΟΛΛΑ LAYERS
def quantum_model(x, params):
    n_qubits = N_QUBITS
    n_layers = N_LAYERS

    expected_params = 2 * n_qubits * n_layers

    if len(params) != expected_params:
        raise ValueError(
            f"Λάθος πλήθος παραμέτρων. "
            f"Αναμένονται {expected_params}, αλλά δόθηκαν {len(params)}."
        )

    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου
    for q in range(n_qubits):
        qc.ry(x, q)

    # Παραμετροποιημένα layers
    param_index = 0

    for layer in range(n_layers):

        # Παραμετροποιημένες περιστροφές σε κάθε qubit
        for q in range(n_qubits):
            qc.ry(params[param_index], q)
            param_index += 1

            qc.rz(params[param_index], q)
            param_index += 1

        # Εμπλοκή με CNOT σε αλυσίδα: q0 -> q1 -> ... -> q(n-1)
        for q in range(n_qubits - 1):
            qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο τελευταίο qubit q(n_qubits - 1)
    observable = SparsePauliOp("Z" + "I" * (n_qubits - 1))

    expectation = np.real(state.expectation_value(observable))

    return expectation


# ΣΥΝΑΡΤΗΣΗ ΜΕΤΑΤΟΠΙΣΗΣ ΠΑΡΑΜΕΤΡΟΥ
def parameter_shift_derivative(model, x, params, index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shift = np.pi / 2

    shifted_plus[index] += shift
    shifted_minus[index] -= shift

    y_plus = model(x, shifted_plus)
    y_minus = model(x, shifted_minus)

    derivative = 0.5 * (y_plus - y_minus)

    return derivative


# ΣΥΝΑΡΤΗΣΗ ΕΝΗΜΕΡΩΣΗΣ ΠΑΡΑΜΕΤΡΩΝ
def gradient_descent_update(params, gradients, learning_rate):
    params = np.array(params)
    gradients = np.array(gradients)

    updated_params = params - learning_rate * gradients

    return updated_params


# ΣΥΝΑΡΤΗΣΗ ΚΟΣΤΟΥΣ
def loss_function(model, x_data, y_data, params):
    total_loss = 0

    for x, y in zip(x_data, y_data):
        y_pred = model(x, params)
        total_loss += (y - y_pred) ** 2

    return total_loss / len(x_data)


# ΥΠΟΛΟΓΙΣΜΟΣ ΤΟΥ GRADIENT ΤΗΣ LOSS FUNCTION
def loss_gradient(model, x_data, y_data, params):
    gradients = np.zeros_like(params)

    for k in range(len(params)):
        grad_sum = 0

        for x, y in zip(x_data, y_data):
            y_pred = model(x, params)

            dy_dtheta = parameter_shift_derivative(
                model, x, params, k
            )

            grad_sum += 2 * (y_pred - y) * dy_dtheta

        gradients[k] = grad_sum / len(x_data)

    return gradients


# ΔΙΑΔΙΚΑΣΙΑ ΕΚΠΑΙΔΕΥΣΗΣ ΜΕ GRADIENT DESCENT
def train_gradient_descent(model, x_data, y_data, params, learning_rate, epochs):
    loss_history = []

    for epoch in range(epochs):
        current_loss = loss_function(model, x_data, y_data, params)
        loss_history.append(current_loss)

        gradients = loss_gradient(model, x_data, y_data, params)

        params = gradient_descent_update(
            params, gradients, learning_rate
        )

        if epoch % 10 == 0:
            print("Epoch:", epoch, "Loss:", current_loss)

    return params, loss_history


# ΠΑΡΑΔΕΙΓΜΑ ΕΚΤΕΛΕΣΗΣ ΤΗΣ ΕΚΠΑΙΔΕΥΣΗΣ ΓΙΑ 2 QUBITS ΚΑΙ 2 LAYERS
np.random.seed(42)

x_data = np.linspace(0, np.pi, 20)
y_data = np.sin(x_data)

n_qubits = N_QUBITS
n_layers = N_LAYERS

# 2 παράμετροι ανά qubit ανά layer: Ry και Rz
params = np.random.uniform(
    0,
    2 * np.pi,
    size=2 * n_qubits * n_layers
)

learning_rate = 0.05
epochs = 300

trained_params, loss_history = train_gradient_descent(
    quantum_model,
    x_data,
    y_data,
    params,
    learning_rate,
    epochs
)

print("Τελικές παράμετροι για 2 qubits με 2 layers:", trained_params)
print("Τελικό loss για 2 qubits με 2 layers:", loss_history[-1])
```

```python
# ΚΩΔΙΚΑΣ 4.7.2 - 2 QUBITS ΜΕ 2 VARIATIONAL LAYERS

import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp
import matplotlib.pyplot as plt


# ΣΤΑΘΕΡΕΣ ΜΟΝΤΕΛΟΥ
N_QUBITS = 2
N_LAYERS = 2


# ΚΒΑΝΤΙΚΟ ΜΟΝΤΕΛΟ ΠΟΛΛΩΝ QUBITS ΜΕ ΠΟΛΛΑ LAYERS
def quantum_model(x, params):
    n_qubits = N_QUBITS
    n_layers = N_LAYERS

    expected_params = 2 * n_qubits * n_layers

    if len(params) != expected_params:
        raise ValueError(
            f"Λάθος πλήθος παραμέτρων. "
            f"Αναμένονται {expected_params}, αλλά δόθηκαν {len(params)}."
        )

    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου
    for q in range(n_qubits):
        qc.ry(x, q)

    # Παραμετροποιημένα layers
    param_index = 0

    for layer in range(n_layers):

        # Παραμετροποιημένες περιστροφές σε κάθε qubit
        for q in range(n_qubits):
            qc.ry(params[param_index], q)
            param_index += 1

            qc.rz(params[param_index], q)
            param_index += 1

        # Εμπλοκή με CNOT σε αλυσίδα: q0 -> q1 -> ... -> q(n-1)
        for q in range(n_qubits - 1):
            qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο τελευταίο qubit q(n_qubits - 1)
    observable = SparsePauliOp("Z" + "I" * (n_qubits - 1))

    expectation = np.real(state.expectation_value(observable))

    return expectation


# ΣΥΝΑΡΤΗΣΗ ΜΕΤΑΤΟΠΙΣΗΣ ΠΑΡΑΜΕΤΡΟΥ
def parameter_shift_derivative(model, x, params, index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shift = np.pi / 2

    shifted_plus[index] += shift
    shifted_minus[index] -= shift

    y_plus = model(x, shifted_plus)
    y_minus = model(x, shifted_minus)

    derivative = 0.5 * (y_plus - y_minus)

    return derivative


# ΣΥΝΑΡΤΗΣΗ ΕΝΗΜΕΡΩΣΗΣ ΠΑΡΑΜΕΤΡΩΝ
def gradient_descent_update(params, gradients, learning_rate):
    params = np.array(params)
    gradients = np.array(gradients)

    updated_params = params - learning_rate * gradients

    return updated_params


# ΣΥΝΑΡΤΗΣΗ ΚΟΣΤΟΥΣ
def loss_function(model, x_data, y_data, params):
    total_loss = 0

    for x, y in zip(x_data, y_data):
        y_pred = model(x, params)
        total_loss += (y - y_pred) ** 2

    return total_loss / len(x_data)


# ΥΠΟΛΟΓΙΣΜΟΣ ΤΟΥ GRADIENT ΤΗΣ LOSS FUNCTION
def loss_gradient(model, x_data, y_data, params):
    gradients = np.zeros_like(params)

    for k in range(len(params)):
        grad_sum = 0

        for x, y in zip(x_data, y_data):
            y_pred = model(x, params)

            dy_dtheta = parameter_shift_derivative(
                model, x, params, k
            )

            grad_sum += 2 * (y_pred - y) * dy_dtheta

        gradients[k] = grad_sum / len(x_data)

    return gradients


# ΔΙΑΔΙΚΑΣΙΑ ΕΚΠΑΙΔΕΥΣΗΣ ΜΕ GRADIENT DESCENT
def train_gradient_descent(model, x_data, y_data, params, learning_rate, epochs):
    loss_history = []

    for epoch in range(epochs):
        current_loss = loss_function(model, x_data, y_data, params)
        loss_history.append(current_loss)

        gradients = loss_gradient(model, x_data, y_data, params)

        params = gradient_descent_update(
            params, gradients, learning_rate
        )

        if epoch % 10 == 0:
            print("Epoch:", epoch, "Loss:", current_loss)

    return params, loss_history


# ΠΑΡΑΔΕΙΓΜΑ ΕΚΤΕΛΕΣΗΣ ΤΗΣ ΕΚΠΑΙΔΕΥΣΗΣ ΓΙΑ 2 QUBITS ΚΑΙ 2 LAYERS
np.random.seed(42)

x_data = np.linspace(0, np.pi, 20)
y_data = np.sin(x_data)

n_qubits = N_QUBITS
n_layers = N_LAYERS

# 2 παράμετροι ανά qubit ανά layer: Ry και Rz
params_2q = np.random.uniform(
    0,
    2 * np.pi,
    size=2 * n_qubits * n_layers
)

learning_rate = 0.05
epochs = 300

trained_params_2q, loss_history_2q = train_gradient_descent(
    quantum_model,
    x_data,
    y_data,
    params_2q,
    learning_rate,
    epochs
)


# ΥΠΟΛΟΓΙΣΜΟΣ ΤΕΛΙΚΩΝ ΠΡΟΒΛΕΨΕΩΝ ΓΙΑ 2 QUBITS
y_pred_2q = np.array([
    quantum_model(x, trained_params_2q) for x in x_data
])

print("Τελικές παράμετροι για 2 qubits με 2 layers:", trained_params_2q)
print("Τελικό loss για 2 qubits με 2 layers:", loss_history_2q[-1])


# ΓΡΑΦΗΜΑ ΕΞΕΛΙΞΗΣ ΤΗΣ ΣΥΝΑΡΤΗΣΗΣ ΚΟΣΤΟΥΣ ΓΙΑ 2 QUBITS
plt.figure(figsize=(8, 5))
plt.plot(loss_history_2q, label="2 qubits - 2 layers")

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss κατά την εκπαίδευση με 2 Qubits και 2 Layers")
plt.legend()
plt.grid(True)
plt.show()


# ΣΥΓΚΡΙΣΗ ΠΡΑΓΜΑΤΙΚΗΣ ΣΥΝΑΡΤΗΣΗΣ ΚΑΙ ΠΡΟΒΛΕΨΗΣ ΓΙΑ 2 QUBITS
plt.figure(figsize=(8, 5))

plt.plot(x_data, y_data, "o-", label="Πραγματική sin(x)")
plt.plot(x_data, y_pred_2q, "x--", label="Πρόβλεψη για 2 qubits - 2 layers")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Προσέγγιση της sin(x) με 2 Qubits και 2 Layers")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ΚΩΔΙΚΑΣ 5.4.2.3 ΚΩΔΙΚΟΠΟΙΗΣΗ ΕΙΣΟΔΟΥ
from qiskit import QuantumCircuit

def encode_input(qc, x, n_qubits):
    """
    Κωδικοποίηση της εισόδου x στο κβαντικό κύκλωμα.
    Η είσοδος μετασχηματίζεται μέσω της feature map φ(x)=2x
    και εφαρμόζεται σε όλα τα qubits μέσω της πύλης Ry(2x).
    """
    phi_x = 2 * x
    
    for q in range(n_qubits):
        qc.ry(phi_x, q)
    
    return qc
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.1
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp
import numpy as np

def quantum_model(x, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου μέσω Ry(2x)
    encode_input(qc, x, n_qubits)

    # Πλήθος variational layers
    if n_qubits == 1:
        n_layers = 1
    else:
        n_layers = 2

    expected_params = 2 * n_qubits * n_layers

    if len(params) != expected_params:
        raise ValueError(
            f"Λάθος πλήθος παραμέτρων. Αναμένονται {expected_params}, "
            f"αλλά δόθηκαν {len(params)}."
        )

    param_index = 0

    # Variational layers
    for layer in range(n_layers):
        # Παραμετροποιημένες περιστροφές Ry(θ), Rz(θ) σε κάθε qubit
        for q in range(n_qubits):
            qc.ry(params[param_index], q)
            param_index += 1

            qc.rz(params[param_index], q)
            param_index += 1
        # Αλυσιδωτή εμπλοκή CNOT για n > 1
        if n_qubits > 1:
            for q in range(n_qubits - 1):
                qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z
    # Για 1 qubit: μέτρηση στο μοναδικό qubit
    # Για n > 1: μέτρηση στο τελευταίο qubit q(n-1)
    observable = SparsePauliOp("Z" + "I" * (n_qubits - 1))

    # Αναμενόμενη τιμή
    output = np.real(state.expectation_value(observable))

    return output
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.2
def loss_function(params, x_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model(x, params, n_qubits) for x in x_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss
```

```python
# ΚΩΔΙΚΑΣ 5.4.3.3 - Εκπαίδευση με Random Search

def train_random_search(x_train, y_train, n_qubits, 
                        iterations=1000, step_size=0.1):
    
    # Πλήθος variational layers
    # Για 1 qubit χρησιμοποιείται η αρχική απλή δομή.
    # Για n > 1 χρησιμοποιούνται 2 variational layers.
    if n_qubits == 1:
        n_layers = 1
    else:
        n_layers = 2

    # Πλήθος εκπαιδεύσιμων παραμέτρων
    n_params = 2 * n_qubits * n_layers

    # Αρχικοποίηση παραμέτρων στο διάστημα [0, 2π]
    params = np.random.uniform(0, 2 * np.pi, size=n_params)

    best_params = params.copy()
    best_loss = loss_function(best_params, x_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):
        # Τυχαία μεταβολή όλων των παραμέτρων
        candidate_params = best_params + step_size * (
            np.random.rand(n_params) - 0.5
        )

        candidate_loss = loss_function(
            candidate_params, x_train, y_train, n_qubits
        )

        # Κρατάμε τις νέες παραμέτρους μόνο αν μειώνουν το loss
        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.4
def parameter_shift_derivative(x, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2

    forward = quantum_model(x, shifted_plus, n_qubits)
    backward = quantum_model(x, shifted_minus, n_qubits)

    derivative = 0.5 * (forward - backward)

    return derivative


def compute_gradient(params, x_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    for k in range(len(params)):
        grad_k = 0

        for x, y in zip(x_data, y_data):
            y_pred = quantum_model(x, params, n_qubits)
            dy_dtheta = parameter_shift_derivative(
                x, params, n_qubits, k
            )

            grad_k += 2 * (y_pred - y) * dy_dtheta

        gradient[k] = grad_k / len(x_data)

    return gradient
```

```python
# ΚΩΔΙΚΑΣ 5.4.3.5 - Εκπαίδευση με Parameter-Shift και Gradient Descent

def train_gradient_descent(x_train, y_train, n_qubits,
                           epochs=None, learning_rate=None):

    # Πλήθος variational layers
    # Για 1 qubit χρησιμοποιείται η αρχική απλή δομή.
    # Για n > 1 χρησιμοποιούνται 2 variational layers.
    if n_qubits == 1:
        n_layers = 1
    else:
        n_layers = 2

    # Πλήθος εκπαιδεύσιμων παραμέτρων
    n_params = 2 * n_qubits * n_layers

    # Προεπιλεγμένες ρυθμίσεις εκπαίδευσης
    if epochs is None:
        if n_qubits == 1:
            epochs = 100
        else:
            epochs = 300

    if learning_rate is None:
        if n_qubits == 10:
            learning_rate = 0.03
        elif n_qubits in [2, 4]:
            learning_rate = 0.05
        else:
            learning_rate = 0.1

    # Αρχικοποίηση παραμέτρων στο διάστημα [0, 2π]
    params = np.random.uniform(0, 2 * np.pi, size=n_params)

    loss_history = []

    for epoch in range(epochs):
        # Υπολογισμός συνάρτησης κόστους
        loss = loss_function(params, x_train, y_train, n_qubits)
        loss_history.append(loss)

        # Υπολογισμός gradient με parameter-shift
        gradient = compute_gradient(params, x_train, y_train, n_qubits)

        # Ενημέρωση παραμέτρων με gradient descent
        params = params - learning_rate * gradient

    return params, loss_history
```

```python
# ΚΩΔΙΚΑΣ 5.4.3.6
# Εκπαίδευση όλων των QML μοντέλων για το πρόβλημα y = sin(2x)

import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp


# ------------------------------------------------------------
# 1. Δημιουργία δεδομένων
# ------------------------------------------------------------

seed = 42
np.random.seed(seed)

N = 40

# Είσοδος στο διάστημα [0, π]
x_data = np.linspace(0, np.pi, N)

# Συνάρτηση στόχος
y_data = np.sin(2 * x_data)

# Κωδικοποιημένη γωνία φ(x)=2x
x_encoded = 2 * x_data

# Τυχαία αναδιάταξη δειγμάτων
indices = np.arange(N)
np.random.shuffle(indices)

# Train-test split
train_size = int(0.75 * N)

train_indices = indices[:train_size]
test_indices = indices[train_size:]

x_train = x_data[train_indices]
y_train = y_data[train_indices]

x_test = x_data[test_indices]
y_test = y_data[test_indices]

print("Πλήθος δειγμάτων εκπαίδευσης:", len(x_train))
print("Πλήθος δειγμάτων ελέγχου:", len(x_test))
print("Εύρος εισόδου x:", (x_data.min(), x_data.max()))
print("Εύρος κωδικοποιημένης γωνίας φ(x)=2x:", (x_encoded.min(), x_encoded.max()))
print("Εύρος εξόδου y=sin(2x):", (y_data.min(), y_data.max()))


# ------------------------------------------------------------
# 2. Κοινές πειραματικές ρυθμίσεις
# ------------------------------------------------------------

qubit_list = [1, 2, 4, 10]

random_iterations = 1000
random_step = 0.1

gradient_learning_rates = {
    1: 0.1,
    2: 0.05,
    4: 0.05,
    10: 0.03
}

gradient_epochs = {
    1: 100,
    2: 300,
    4: 300,
    10: 300
}


def get_num_layers(n_qubits):
    """
    Επιστρέφει το πλήθος των variational layers.
    Για 1 qubit χρησιμοποιείται η αρχική απλή δομή.
    Για n > 1 χρησιμοποιούνται 2 variational layers.
    """
    if n_qubits == 1:
        return 1
    return 2


def get_num_params(n_qubits):
    """
    Υπολογισμός πλήθους εκπαιδεύσιμων παραμέτρων.
    m = 2nL
    """
    n_layers = get_num_layers(n_qubits)
    return 2 * n_qubits * n_layers


def initialize_params(n_qubits):
    """
    Τυχαία αρχικοποίηση παραμέτρων στο διάστημα [0, 2π].
    """
    n_params = get_num_params(n_qubits)
    return np.random.uniform(0, 2 * np.pi, size=n_params)


def get_output_observable(n_qubits):
    """
    Observable εξόδου.

    Στο Qiskit, το αριστερότερο γράμμα του Pauli string
    αντιστοιχεί στο qubit με τον μεγαλύτερο δείκτη.

    Επομένως, το "Z" + "I"*(n_qubits-1) αντιστοιχεί
    σε μέτρηση Pauli-Z στο τελευταίο qubit q(n-1).
    """
    return SparsePauliOp("Z" + "I" * (n_qubits - 1))


# ------------------------------------------------------------
# 3. Κωδικοποίηση εισόδου
# ------------------------------------------------------------

def encode_input(qc, x, n_qubits):
    """
    Κωδικοποίηση της εισόδου x στο κβαντικό κύκλωμα.
    Η είσοδος μετασχηματίζεται μέσω της feature map φ(x)=2x
    και εφαρμόζεται σε όλα τα qubits μέσω της πύλης Ry(2x).
    """
    phi_x = 2 * x

    for q in range(n_qubits):
        qc.ry(phi_x, q)

    return qc


# ------------------------------------------------------------
# 4. Ορισμός QML μοντέλου
# ------------------------------------------------------------

def quantum_model(x, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου μέσω Ry(2x)
    encode_input(qc, x, n_qubits)

    # Πλήθος variational layers
    n_layers = get_num_layers(n_qubits)

    # Έλεγχος πλήθους παραμέτρων
    expected_params = get_num_params(n_qubits)

    if len(params) != expected_params:
        raise ValueError(
            f"Λάθος πλήθος παραμέτρων. Αναμένονται {expected_params}, "
            f"αλλά δόθηκαν {len(params)}."
        )

    param_index = 0

    # Variational layers
    for layer in range(n_layers):

        # Παραμετροποιημένες περιστροφές Ry(θ), Rz(θ) σε κάθε qubit
        for q in range(n_qubits):
            qc.ry(params[param_index], q)
            param_index += 1

            qc.rz(params[param_index], q)
            param_index += 1

        # Αλυσιδωτή εμπλοκή CNOT για n > 1
        if n_qubits > 1:
            for q in range(n_qubits - 1):
                qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z
    # Για 1 qubit: μέτρηση στο μοναδικό qubit
    # Για n > 1: μέτρηση στο τελευταίο qubit q(n-1)
    observable = get_output_observable(n_qubits)

    # Αναμενόμενη τιμή
    output = np.real(state.expectation_value(observable))

    return output


# ------------------------------------------------------------
# 5. Συνάρτηση κόστους και προβλέψεις
# ------------------------------------------------------------

def loss_function(params, x_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model(x, params, n_qubits) for x in x_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss


def predict_model(params, x_data, n_qubits):
    predictions = np.array([
        quantum_model(x, params, n_qubits) for x in x_data
    ])

    return predictions


# ------------------------------------------------------------
# 6. Εκπαίδευση με Random Search
# ------------------------------------------------------------

def train_random_search(x_train, y_train, n_qubits,
                        iterations=1000, step_size=0.1,
                        initial_params=None):

    n_params = get_num_params(n_qubits)

    # Αρχικοποίηση παραμέτρων
    if initial_params is None:
        params = initialize_params(n_qubits)
    else:
        params = initial_params.copy()

    best_params = params.copy()
    best_loss = loss_function(best_params, x_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):

        # Τυχαία μεταβολή όλων των παραμέτρων
        candidate_params = best_params + step_size * (
            np.random.rand(n_params) - 0.5
        )

        candidate_loss = loss_function(
            candidate_params,
            x_train,
            y_train,
            n_qubits
        )

        # Κρατάμε τις νέες παραμέτρους μόνο αν μειώνουν το loss
        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history


# ------------------------------------------------------------
# 7. Parameter-shift και Gradient Descent
# ------------------------------------------------------------

def parameter_shift_derivative(x, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    # Μετατόπιση της παραμέτρου κατά +π/2 και -π/2
    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2

    # Εκτέλεση του μοντέλου για τις δύο μετατοπισμένες τιμές
    forward = quantum_model(x, shifted_plus, n_qubits)
    backward = quantum_model(x, shifted_minus, n_qubits)

    # Κανόνας parameter-shift
    derivative = 0.5 * (forward - backward)

    return derivative


def compute_gradient(params, x_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    # Υπολογισμός παραγώγου της συνάρτησης κόστους
    # ως προς κάθε εκπαιδεύσιμη παράμετρο
    for k in range(len(params)):
        grad_k = 0

        for x, y in zip(x_data, y_data):
            y_pred = quantum_model(x, params, n_qubits)

            dy_dtheta = parameter_shift_derivative(
                x,
                params,
                n_qubits,
                k
            )

            grad_k += 2 * (y_pred - y) * dy_dtheta

        gradient[k] = grad_k / len(x_data)

    return gradient


def train_gradient_descent(x_train, y_train, n_qubits,
                           epochs=None, learning_rate=None,
                           initial_params=None):

    # Προεπιλεγμένες ρυθμίσεις εκπαίδευσης
    if epochs is None:
        epochs = gradient_epochs[n_qubits]

    if learning_rate is None:
        learning_rate = gradient_learning_rates[n_qubits]

    # Αρχικοποίηση παραμέτρων
    if initial_params is None:
        params = initialize_params(n_qubits)
    else:
        params = initial_params.copy()

    loss_history = []

    for epoch in range(epochs):

        # Υπολογισμός συνάρτησης κόστους
        loss = loss_function(params, x_train, y_train, n_qubits)
        loss_history.append(loss)

        # Υπολογισμός gradient με parameter-shift
        gradient = compute_gradient(params, x_train, y_train, n_qubits)

        # Ενημέρωση παραμέτρων με gradient descent
        params = params - learning_rate * gradient

    return params, loss_history


# ------------------------------------------------------------
# 8. Εκπαίδευση όλων των μοντέλων
# ------------------------------------------------------------

results_regression = {}
summary_regression = []

for n_qubits in qubit_list:

    print("\n" + "=" * 60)
    print("Μοντέλο", n_qubits, "qubits")
    print("=" * 60)

    n_layers = get_num_layers(n_qubits)
    n_params = get_num_params(n_qubits)

    print("Variational layers:", n_layers)
    print("Πλήθος παραμέτρων:", n_params)

    # Ίδιες αρχικές παράμετροι για Random Search και Gradient Descent
    initial_params = initialize_params(n_qubits)

    # -----------------------------
    # Random Search
    # -----------------------------

    print("Εκπαίδευση με Random Search...")

    params_random, loss_random = train_random_search(
        x_train,
        y_train,
        n_qubits,
        iterations=random_iterations,
        step_size=random_step,
        initial_params=initial_params
    )

    train_loss_random = loss_function(
        params_random,
        x_train,
        y_train,
        n_qubits
    )

    test_loss_random = loss_function(
        params_random,
        x_test,
        y_test,
        n_qubits
    )

    results_regression[(n_qubits, "random_search")] = {
        "params": params_random,
        "loss_history": loss_random,
        "train_loss": train_loss_random,
        "test_loss": test_loss_random,
        "n_layers": n_layers,
        "n_params": n_params
    }

    summary_regression.append([
        n_qubits,
        n_layers,
        n_params,
        "Random Search",
        train_loss_random,
        test_loss_random
    ])

    print("Train loss Random Search:", train_loss_random)
    print("Test loss Random Search:", test_loss_random)

    # -----------------------------
    # Parameter-shift + Gradient Descent
    # -----------------------------

    print("Εκπαίδευση με Parameter-shift + Gradient Descent...")

    params_gd, loss_gd = train_gradient_descent(
        x_train,
        y_train,
        n_qubits,
        epochs=gradient_epochs[n_qubits],
        learning_rate=gradient_learning_rates[n_qubits],
        initial_params=initial_params
    )

    train_loss_gd = loss_function(
        params_gd,
        x_train,
        y_train,
        n_qubits
    )

    test_loss_gd = loss_function(
        params_gd,
        x_test,
        y_test,
        n_qubits
    )

    results_regression[(n_qubits, "parameter_shift_gd")] = {
        "params": params_gd,
        "loss_history": loss_gd,
        "train_loss": train_loss_gd,
        "test_loss": test_loss_gd,
        "n_layers": n_layers,
        "n_params": n_params
    }

    summary_regression.append([
        n_qubits,
        n_layers,
        n_params,
        "Parameter-shift + GD",
        train_loss_gd,
        test_loss_gd
    ])

    print("Train loss Parameter-shift + GD:", train_loss_gd)
    print("Test loss Parameter-shift + GD:", test_loss_gd)


# ------------------------------------------------------------
# 9. Συγκεντρωτική εμφάνιση αποτελεσμάτων
# ------------------------------------------------------------

print("\n\nΣΥΓΚΕΝΤΡΩΤΙΚΑ ΑΠΟΤΕΛΕΣΜΑΤΑ")
print("-" * 110)
print("Qubits | Layers | Παράμετροι | Μέθοδος Εκπαίδευσης | Train Loss | Test Loss")
print("-" * 110)

for row in summary_regression:
    n_qubits, n_layers, n_params, method, train_loss, test_loss = row

    print(
        n_qubits, "|",
        n_layers, "|",
        n_params, "|",
        method, "|",
        train_loss, "|",
        test_loss
    )
```

```python
#ΚΩΔΙΚΑΣ 5.5.2.3
from qiskit import QuantumCircuit

def encode_iris_input(qc, x_vec, n_qubits):
    """
    Κωδικοποίηση δύο χαρακτηριστικών Iris στο κβαντικό κύκλωμα.

    x_vec[0] -> petal length
    x_vec[1] -> petal width
    """

    for q in range(n_qubits):
        qc.ry(x_vec[0], q)
        qc.rz(x_vec[1], q)

    return qc
```

```python
# ΚΩΔΙΚΑΣ 5.5.3.1
# Ορισμός QML μοντέλου για το Iris dataset
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

def get_num_layers(n_qubits):
 
    if n_qubits == 1:
        return 1
    return 2

def get_num_params(n_qubits):
   
    n_layers = get_num_layers(n_qubits)
    return 2 * n_qubits * n_layers


def get_output_observable(n_qubits):
 
    return SparsePauliOp("Z" + "I" * (n_qubits - 1))

def quantum_model_iris(x_vec, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση των δύο χαρακτηριστικών Iris
    encode_iris_input(qc, x_vec, n_qubits)

    # Πλήθος variational layers
    n_layers = get_num_layers(n_qubits)

    # Έλεγχος πλήθους παραμέτρων
    expected_params = get_num_params(n_qubits)

    if len(params) != expected_params:
        raise ValueError(
            f"Λάθος πλήθος παραμέτρων. Αναμένονται {expected_params}, "
            f"αλλά δόθηκαν {len(params)}."
        )

    param_index = 0

    # Variational layers
    for layer in range(n_layers):

        # Παραμετροποιημένες περιστροφές Ry(θ), Rz(θ) σε κάθε qubit
        for q in range(n_qubits):
            qc.ry(params[param_index], q)
            param_index += 1

            qc.rz(params[param_index], q)
            param_index += 1

        # Αλυσιδωτή εμπλοκή CNOT για n > 1
        if n_qubits > 1:
            for q in range(n_qubits - 1):
                qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z
    # Για 1 qubit: μέτρηση στο μοναδικό qubit
    # Για n > 1: μέτρηση στο τελευταίο qubit q(n-1)
    observable = get_output_observable(n_qubits)

    # Αναμενόμενη τιμή
    output = np.real(state.expectation_value(observable))

    return output
```

```python
# ΚΩΔΙΚΑΣ 5.5.3.2
# Συνάρτηση κόστους, προβλέψεις και ακρίβεια ταξινόμησης για το Iris dataset

def iris_loss_function(params, X_data, y_data, n_qubits):
    # Υπολογισμός συνεχών προβλέψεων στο διάστημα [-1, 1]
    predictions = np.array([
        quantum_model_iris(x, params, n_qubits) for x in X_data
    ])
    # Μέσο τετραγωνικό σφάλμα ως προς τις ετικέτες {-1, +1}
    loss = np.mean((predictions - y_data) ** 2)

    return loss

def iris_predict_values(params, X_data, n_qubits):
    # Συνεχείς έξοδοι του QML μοντέλου
    predictions = np.array([
        quantum_model_iris(x, params, n_qubits) for x in X_data
    ])

    return predictions

def iris_predict_classes(params, X_data, n_qubits):
    predictions = iris_predict_values(params, X_data, n_qubits)

    # Κανόνας απόφασης με threshold στο 0
    # prediction < 0  -> κλάση 0, Setosa
    # prediction >= 0 -> κλάση 1, Versicolor
    predicted_classes = np.where(predictions >= 0, 1, 0)

    return predicted_classes

def iris_accuracy(params, X_data, y_class, n_qubits):
    predicted_classes = iris_predict_classes(params, X_data, n_qubits)

    # Ποσοστό σωστών ταξινομήσεων
    accuracy = np.mean(predicted_classes == y_class)

    return accuracy
```

```python
# ΚΩΔΙΚΑΣ 5.5.3.3
# Εκπαίδευση του QML μοντέλου Iris με Random Search
def train_iris_random_search(X_train, y_train, n_qubits,
                             iterations=500, step_size=0.1,
                             initial_params=None):
    # Πλήθος εκπαιδεύσιμων παραμέτρων
    n_params = get_num_params(n_qubits)

    # Αρχικοποίηση παραμέτρων
    if initial_params is None:
        params = np.random.uniform(0, 2 * np.pi, size=n_params)
    else:
        params = initial_params.copy()

    best_params = params.copy()
    best_loss = iris_loss_function(best_params, X_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):

        # Τυχαία μεταβολή όλων των παραμέτρων
        candidate_params = best_params + step_size * (
            np.random.rand(n_params) - 0.5
        )

        candidate_loss = iris_loss_function(
            candidate_params,
            X_train,
            y_train,
            n_qubits
        )
        # Κρατάμε τις νέες παραμέτρους μόνο αν μειώνουν το loss
        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history
```

```python
# ΚΩΔΙΚΑΣ 5.5.3.4
# Εκπαίδευση του QML μοντέλου Iris με Parameter-Shift και Gradient Descent
def iris_parameter_shift_derivative(x_vec, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()
    # Μετατόπιση της παραμέτρου κατά +π/2 και -π/2
    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2
    # Εκτέλεση του μοντέλου για τις δύο μετατοπισμένες τιμές
    forward = quantum_model_iris(x_vec, shifted_plus, n_qubits)
    backward = quantum_model_iris(x_vec, shifted_minus, n_qubits)
    # Κανόνας parameter-shift
    derivative = 0.5 * (forward - backward)
    return derivative

def iris_compute_gradient(params, X_data, y_data, n_qubits):
    gradient = np.zeros_like(params)
    # Υπολογισμός gradient ως προς κάθε εκπαιδεύσιμη παράμετρο
    for k in range(len(params)):
        grad_k = 0
        for x_vec, y in zip(X_data, y_data):
            y_pred = quantum_model_iris(x_vec, params, n_qubits)

            dy_dtheta = iris_parameter_shift_derivative(
                x_vec,
                params,
                n_qubits,
                k
            )
            grad_k += 2 * (y_pred - y) * dy_dtheta
        gradient[k] = grad_k / len(X_data)
    return gradient

def train_iris_gradient_descent(X_train, y_train, n_qubits,
                                epochs=50, learning_rate=0.1,
                                initial_params=None):
    # Πλήθος εκπαιδεύσιμων παραμέτρων
    n_params = get_num_params(n_qubits)

    # Αρχικοποίηση παραμέτρων
    if initial_params is None:
        params = np.random.uniform(0, 2 * np.pi, size=n_params)
    else:
        params = initial_params.copy()
    loss_history = []
    for epoch in range(epochs):
        # Υπολογισμός συνάρτησης κόστους
        loss = iris_loss_function(params, X_train, y_train, n_qubits)
        loss_history.append(loss)
        # Υπολογισμός gradient με parameter-shift
        gradient = iris_compute_gradient(params, X_train, y_train, n_qubits)
        # Ενημέρωση παραμέτρων με gradient descent
        params = params - learning_rate * gradient
    return params, loss_history
```

```python
#ΚΩΔΙΚΑΣ 5.6.2.3
from qiskit import QuantumCircuit

def encode_blobs_input(qc, x_vec, n_qubits):
    """
    Κωδικοποίηση δύο χαρακτηριστικών synthetic blobs στο κβαντικό κύκλωμα.

    x_vec[0] -> πρώτη συντεταγμένη
    x_vec[1] -> δεύτερη συντεταγμένη
    """

    for q in range(n_qubits):
        qc.ry(x_vec[0], q)
        qc.rz(x_vec[1], q)

    return qc
```

```python
# ΚΩΔΙΚΑΣ 5.6.3.1
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

def quantum_model_blobs(x_vec, params, n_qubits):
    qc = QuantumCircuit(n_qubits)
    # Κωδικοποίηση των δύο χαρακτηριστικών synthetic blobs
    encode_blobs_input(qc, x_vec, n_qubits)
    # Πλήθος variational layers
    n_layers = get_num_layers(n_qubits)
    # Έλεγχος πλήθους παραμέτρων
    expected_params = get_num_params(n_qubits)

    if len(params) != expected_params:
        raise ValueError(
            f"Λάθος πλήθος παραμέτρων. Αναμένονται {expected_params}, "
            f"αλλά δόθηκαν {len(params)}."
        )
    param_index = 0

    # Variational layers
    for layer in range(n_layers):

        # Παραμετροποιημένες περιστροφές Ry(θ), Rz(θ) σε κάθε qubit
        for q in range(n_qubits):
            qc.ry(params[param_index], q)
            param_index += 1
            qc.rz(params[param_index], q)
            param_index += 1

        # Αλυσιδωτή εμπλοκή CNOT για n > 1
        if n_qubits > 1:
            for q in range(n_qubits - 1):
                qc.cx(q, q + 1)
    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)
    # Μέτρηση Pauli-Z
    observable = get_output_observable(n_qubits)
    # Αναμενόμενη τιμή
    output = np.real(state.expectation_value(observable))

    return output
```

```python
#ΚΩΔΙΚΑΣ 5.6.3.2
def blobs_loss_function(params, X_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model_blobs(x, params, n_qubits) for x in X_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss


def blobs_predict_values(params, X_data, n_qubits):
    predictions = np.array([
        quantum_model_blobs(x, params, n_qubits) for x in X_data
    ])

    return predictions


def blobs_predict_classes(params, X_data, n_qubits):
    predictions = blobs_predict_values(params, X_data, n_qubits)

    predicted_classes = np.where(predictions >= 0, 1, 0)

    return predicted_classes


def blobs_accuracy(params, X_data, y_class, n_qubits):
    predicted_classes = blobs_predict_classes(params, X_data, n_qubits)

    accuracy = np.mean(predicted_classes == y_class)

    return accuracy
```

```python
# ΚΩΔΙΚΑΣ 5.6.3.3
# Εκπαίδευση του QML μοντέλου blobs με Random Search
def train_blobs_random_search(X_train, y_train, n_qubits,
                              iterations=500, step_size=0.1,
                              initial_params=None):

    # Πλήθος εκπαιδεύσιμων παραμέτρων
    n_params = get_num_params(n_qubits)
    # Αρχικοποίηση παραμέτρων
    if initial_params is None:
        params = np.random.uniform(0, 2 * np.pi, size=n_params)
    else:
        params = initial_params.copy()

    best_params = params.copy()
    best_loss = blobs_loss_function(best_params, X_train, y_train, n_qubits)
    loss_history = []

    for i in range(iterations):

        # Τυχαία μεταβολή όλων των παραμέτρων
        candidate_params = best_params + step_size * (
            np.random.rand(n_params) - 0.5
        )

        candidate_loss = blobs_loss_function(
            candidate_params,
            X_train,
            y_train,
            n_qubits
        )
        # Κρατάμε τις νέες παραμέτρους μόνο αν μειώνουν το loss
        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()
        loss_history.append(best_loss)

    return best_params, loss_history
```

```python
# ΚΩΔΙΚΑΣ 5.6.3.4
# Εκπαίδευση του QML μοντέλου blobs με Parameter-Shift και Gradient Descent
def blobs_parameter_shift_derivative(x_vec, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()
    # Μετατόπιση της παραμέτρου κατά +π/2 και -π/2
    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2
    forward = quantum_model_blobs(x_vec, shifted_plus, n_qubits)
    backward = quantum_model_blobs(x_vec, shifted_minus, n_qubits)
    derivative = 0.5 * (forward - backward)

    return derivative

def blobs_compute_gradient(params, X_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    for x_vec, y in zip(X_data, y_data):
        y_pred = quantum_model_blobs(x_vec, params, n_qubits)

        for k in range(len(params)):
            dy_dtheta = blobs_parameter_shift_derivative(
                x_vec,
                params,
                n_qubits,
                k
            )
            gradient[k] += 2 * (y_pred - y) * dy_dtheta
    gradient = gradient / len(X_data)

    return gradient

def train_blobs_gradient_descent(X_train, y_train, n_qubits,
                                 epochs=50, learning_rate=0.1,
                                 initial_params=None):
    # Πλήθος εκπαιδεύσιμων παραμέτρων
    n_params = get_num_params(n_qubits)
    # Αρχικοποίηση παραμέτρων
    if initial_params is None:
        params = np.random.uniform(0, 2 * np.pi, size=n_params)
    else:
        params = initial_params.copy()

    loss_history = []

    for epoch in range(epochs):

        # Υπολογισμός συνάρτησης κόστους
        loss = blobs_loss_function(params, X_train, y_train, n_qubits)
        loss_history.append(loss)

        # Υπολογισμός gradient με parameter-shift
        gradient = blobs_compute_gradient(params, X_train, y_train, n_qubits)

        # Ενημέρωση παραμέτρων με gradient descent
        params = params - learning_rate * gradient

    return params, loss_history
```

```python
# ΕΝΙΑΙΟΣ ΚΩΔΙΚΑΣ 5.6.3
# Πλήθος qubits που εξετάζονται
qubit_list = [1, 2, 4, 10]

# Ρυθμίσεις εκπαίδευσης
blobs_random_iterations = 500
blobs_epochs = 50
blobs_learning_rate = 0.1
blobs_random_step = 0.1

# Αποθήκευση αποτελεσμάτων
results_blobs = {}
summary_blobs = []

for n_qubits in qubit_list:

    print("\n" + "=" * 60)
    print("Synthetic Blobs model:", n_qubits, "qubits")
    print("=" * 60)

    n_layers = get_num_layers(n_qubits)
    n_params = get_num_params(n_qubits)

    print("Variational layers:", n_layers)
    print("Πλήθος παραμέτρων:", n_params)

    # Ίδιες αρχικές παράμετροι για δίκαιη σύγκριση των δύο μεθόδων
    initial_params = np.random.uniform(0, 2 * np.pi, size=n_params)

    # --------------------------------------------------
    # Random Search
    # --------------------------------------------------

    print("Εκπαίδευση με Random Search...")

    params_random, loss_random = train_blobs_random_search(
        X_blobs_train,
        y_blobs_train,
        n_qubits,
        iterations=blobs_random_iterations,
        step_size=blobs_random_step,
        initial_params=initial_params
    )

    train_loss_random = blobs_loss_function(
        params_random,
        X_blobs_train,
        y_blobs_train,
        n_qubits
    )

    test_loss_random = blobs_loss_function(
        params_random,
        X_blobs_test,
        y_blobs_test,
        n_qubits
    )

    train_acc_random = blobs_accuracy(
        params_random,
        X_blobs_train,
        y_blobs_train_class,
        n_qubits
    )

    test_acc_random = blobs_accuracy(
        params_random,
        X_blobs_test,
        y_blobs_test_class,
        n_qubits
    )

    results_blobs[(n_qubits, "random_search")] = {
        "params": params_random,
        "loss_history": loss_random,
        "train_loss": train_loss_random,
        "test_loss": test_loss_random,
        "train_accuracy": train_acc_random,
        "test_accuracy": test_acc_random,
        "n_layers": n_layers,
        "n_params": n_params
    }

    summary_blobs.append([
        n_qubits,
        n_layers,
        n_params,
        "Random Search",
        train_loss_random,
        test_loss_random,
        train_acc_random,
        test_acc_random
    ])

    print("Train loss Random Search:", train_loss_random)
    print("Test loss Random Search:", test_loss_random)
    print("Train accuracy Random Search:", train_acc_random)
    print("Test accuracy Random Search:", test_acc_random)

    # --------------------------------------------------
    # Parameter-shift + Gradient Descent
    # --------------------------------------------------

    print("Εκπαίδευση με Parameter-shift + Gradient Descent...")

    params_gd, loss_gd = train_blobs_gradient_descent(
        X_blobs_train,
        y_blobs_train,
        n_qubits,
        epochs=blobs_epochs,
        learning_rate=blobs_learning_rate,
        initial_params=initial_params
    )

    train_loss_gd = blobs_loss_function(
        params_gd,
        X_blobs_train,
        y_blobs_train,
        n_qubits
    )

    test_loss_gd = blobs_loss_function(
        params_gd,
        X_blobs_test,
        y_blobs_test,
        n_qubits
    )

    train_acc_gd = blobs_accuracy(
        params_gd,
        X_blobs_train,
        y_blobs_train_class,
        n_qubits
    )

    test_acc_gd = blobs_accuracy(
        params_gd,
        X_blobs_test,
        y_blobs_test_class,
        n_qubits
    )

    results_blobs[(n_qubits, "parameter_shift_gd")] = {
        "params": params_gd,
        "loss_history": loss_gd,
        "train_loss": train_loss_gd,
        "test_loss": test_loss_gd,
        "train_accuracy": train_acc_gd,
        "test_accuracy": test_acc_gd,
        "n_layers": n_layers,
        "n_params": n_params
    }

    summary_blobs.append([
        n_qubits,
        n_layers,
        n_params,
        "Parameter-shift + GD",
        train_loss_gd,
        test_loss_gd,
        train_acc_gd,
        test_acc_gd
    ])

    print("Train loss Parameter-shift + GD:", train_loss_gd)
    print("Test loss Parameter-shift + GD:", test_loss_gd)
    print("Train accuracy Parameter-shift + GD:", train_acc_gd)
    print("Test accuracy Parameter-shift + GD:", test_acc_gd)


# --------------------------------------------------
# Συγκεντρωτική εμφάνιση αποτελεσμάτων
# --------------------------------------------------

print("\n\nΣΥΓΚΕΝΤΡΩΤΙΚΑ ΑΠΟΤΕΛΕΣΜΑΤΑ SYNTHETIC BLOBS")
print("-" * 130)
print("Qubits | Layers | Παράμετροι | Μέθοδος | Train Loss | Test Loss | Train Accuracy | Test Accuracy")
print("-" * 130)

for row in summary_blobs:
    n_qubits, n_layers, n_params, method, train_loss, test_loss, train_acc, test_acc = row

    print(
        n_qubits, "|",
        n_layers, "|",
        n_params, "|",
        method, "|",
        train_loss, "|",
        test_loss, "|",
        train_acc, "|",
        test_acc
    )
```

```python
# ΣΧΗΜΑΤΑ 5.6.4.3 - 5.6.4.6
# Περιοχές ταξινόμησης για όλα τα QML μοντέλα Synthetic Blobs

def plot_blobs_decision_regions(n_qubits, grid_resolution=60):
    methods = [
        ("random_search", "Random Search"),
        ("parameter_shift_gd", "Parameter-shift + GD")
    ]

    x1_min, x1_max = 0, np.pi
    x2_min, x2_max = 0, np.pi

    x1_grid = np.linspace(x1_min, x1_max, grid_resolution)
    x2_grid = np.linspace(x2_min, x2_max, grid_resolution)

    xx, yy = np.meshgrid(x1_grid, x2_grid)
    grid_points = np.c_[xx.ravel(), yy.ravel()]

    plt.figure(figsize=(12, 5))

    for index, (method_key, method_name) in enumerate(methods, start=1):
        params = results_blobs[(n_qubits, method_key)]["params"]

        predicted_grid = blobs_predict_classes(
            params,
            grid_points,
            n_qubits
        )

        predicted_grid = predicted_grid.reshape(xx.shape)

        plt.subplot(1, 2, index)

        plt.contourf(
            xx,
            yy,
            predicted_grid,
            levels=[-0.5, 0.5, 1.5],
            alpha=0.25
        )

        # Training samples - Class 0
        plt.scatter(
            X_blobs_train[y_blobs_train_class == 0, 0],
            X_blobs_train[y_blobs_train_class == 0, 1],
            marker="o",
            label="Class 0 - Train"
        )

        # Test samples - Class 0
        plt.scatter(
            X_blobs_test[y_blobs_test_class == 0, 0],
            X_blobs_test[y_blobs_test_class == 0, 1],
            marker="s",
            label="Class 0 - Test"
        )

        # Training samples - Class 1
        plt.scatter(
            X_blobs_train[y_blobs_train_class == 1, 0],
            X_blobs_train[y_blobs_train_class == 1, 1],
            marker="o",
            label="Class 1 - Train"
        )

        # Test samples - Class 1
        plt.scatter(
            X_blobs_test[y_blobs_test_class == 1, 0],
            X_blobs_test[y_blobs_test_class == 1, 1],
            marker="s",
            label="Class 1 - Test"
        )

        plt.xlabel("x1 normalized")
        plt.ylabel("x2 normalized")
        plt.title(str(n_qubits) + " qubits - " + method_name)
        plt.grid(True)

    plt.suptitle(
        "Περιοχές ταξινόμησης Synthetic Blobs - "
        + str(n_qubits)
        + " qubits"
    )

    plt.tight_layout()
    plt.show()
```
