```python
#ΚΩΔΙΚΑΣ 4.5.1
import numpy as np

# ΑΠΛΟ ΠΑΡΑΔΕΙΓΜΑ ΚΒΑΝΤΙΚΟΥ ΜΟΝΤΕΛΟΥ
# Η έξοδος παριστάνει μια αναμενόμενη τιμή, όπως συμβαίνει σε QML μοντέλα.
def model(x, params):
    return np.cos(x + np.sum(params))

# ΣΥΝΑΡΤΗΣΗ ΜΕΤΑΤΟΠΙΣΗΣ ΠΑΡΑΜΕΤΡΟΥ
def parameter_shift_derivative(model, x, params, index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shift = np.pi / 2

    shifted_plus[index] += shift
    shifted_minus[index] -= shift

    y_plus = model(x, shifted_plus)
    y_minus = model(x, shifted_minus)

    derivative = 0.5 * (y_plus - y_minus)

    return derivative

# ΥΠΟΛΟΓΙΣΜΟΣ GRADIENT ΓΙΑ ΟΛΕΣ ΤΙΣ ΠΑΡΑΜΕΤΡΟΥΣ
def parameter_shift_gradient(model, x, params):
    gradients = np.zeros_like(params)

    for index in range(len(params)):
        gradients[index] = parameter_shift_derivative(model, x, params, index)

    return gradients

# ΣΥΝΑΡΤΗΣΗ ΕΝΗΜΕΡΩΣΗΣ ΠΑΡΑΜΕΤΡΩΝ ΑΠΟ ΤΟ 4.4
def gradient_descent_update(params, gradients, learning_rate):
    params = np.array(params)
    gradients = np.array(gradients)

    updated_params = params - learning_rate * gradients
    return updated_params

# ΠΑΡΑΔΕΙΓΜΑ ΕΚΤΕΛΕΣΗΣ
x = 0.5
params = np.array([0.3, 1.2])
learning_rate = 0.1

print("Αρχικές παράμετροι:", params)

y_pred = model(x, params)
print("Έξοδος μοντέλου:", y_pred)

gradients = parameter_shift_gradient(model, x, params)
print("Gradient:", gradients)

updated_params = gradient_descent_update(params, gradients, learning_rate)
print("Νέες παράμετροι:", updated_params)
```

```python
# ΚΩΔΙΚΑΣ 4.6
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

# ΚΒΑΝΤΙΚΟ ΜΟΝΤΕΛΟ ΠΟΛΛΩΝ QUBITS
def quantum_model(x, params):
    n_qubits = len(params) // 2

    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου
    for q in range(n_qubits):
        qc.ry(x, q)

    # Παραμετροποιημένο τμήμα
    for q in range(n_qubits):
        qc.ry(params[q], q)
        qc.rz(params[n_qubits + q], q)

    # Εμπλοκή με CNOT σε αλυσίδα
    for q in range(n_qubits - 1):
        qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο πρώτο qubit q0
    observable = SparsePauliOp("I" * (n_qubits - 1) + "Z")

    expectation = np.real(state.expectation_value(observable))

    return expectation



# ΣΥΝΑΡΤΗΣΗ ΜΕΤΑΤΟΠΙΣΗΣ ΠΑΡΑΜΕΤΡΟΥ
def parameter_shift_derivative(model, x, params, index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shift = np.pi / 2

    shifted_plus[index] += shift
    shifted_minus[index] -= shift

    y_plus = model(x, shifted_plus)
    y_minus = model(x, shifted_minus)

    derivative = 0.5 * (y_plus - y_minus)

    return derivative

# ΥΠΟΛΟΓΙΣΜΟΣ GRADIENT ΓΙΑ ΟΛΕΣ ΤΙΣ ΠΑΡΑΜΕΤΡΟΥΣ
def parameter_shift_gradient(model, x, params):
    gradients = np.zeros_like(params)

    for index in range(len(params)):
        gradients[index] = parameter_shift_derivative(model, x, params, index)

    return gradients

# ΣΥΝΑΡΤΗΣΗ ΕΝΗΜΕΡΩΣΗΣ ΠΑΡΑΜΕΤΡΩΝ ΑΠΟ ΤΟ 4.4
def gradient_descent_update(params, gradients, learning_rate):
    params = np.array(params)
    gradients = np.array(gradients)

    updated_params = params - learning_rate * gradients
    return updated_params




# ΣΥΝΑΡΤΗΣΗ ΚΟΣΤΟΥΣ
def loss_function(model, x_data, y_data, params):
    total_loss = 0

    for x, y in zip(x_data, y_data):
        y_pred = model(x, params)
        total_loss += (y - y_pred) ** 2

    return total_loss / len(x_data)


# ΥΠΟΛΟΓΙΣΜΟΣ ΤΟΥ GRADIENT ΤΗΣ LOSS FUNCTION
def loss_gradient(model, x_data, y_data, params):
    gradients = np.zeros_like(params)

    for k in range(len(params)):
        grad_sum = 0

        for x, y in zip(x_data, y_data):
            y_pred = model(x, params)

            dy_dtheta = parameter_shift_derivative(
                model, x, params, k
            )

            grad_sum += 2 * (y_pred - y) * dy_dtheta

        gradients[k] = grad_sum / len(x_data)

    return gradients


# ΔΙΑΔΙΚΑΣΙΑ ΕΚΠΑΙΔΕΥΣΗΣ ΜΕ GRADIENT DESCENT
def train_gradient_descent(model, x_data, y_data, params, learning_rate, epochs):
    loss_history = []

    for epoch in range(epochs):
        current_loss = loss_function(model, x_data, y_data, params)
        loss_history.append(current_loss)

        gradients = loss_gradient(model, x_data, y_data, params)

        params = gradient_descent_update(
            params, gradients, learning_rate
        )

        if epoch % 10 == 0:
            print("Epoch:", epoch, "Loss:", current_loss)

    return params, loss_history


# ΠΑΡΑΔΕΙΓΜΑ ΕΚΤΕΛΕΣΗΣ ΤΗΣ ΕΚΠΑΙΔΕΥΣΗΣ ΓΙΑ 2 QUBITS
np.random.seed(42)
x_data = np.linspace(0, np.pi, 20)
y_data = np.sin(x_data)

n_qubits = 2
params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)

learning_rate = 0.1
epochs = 100

trained_params, loss_history = train_gradient_descent(
    quantum_model,
    x_data,
    y_data,
    params,
    learning_rate,
    epochs
)

print("Τελικές παράμετροι:", trained_params)
print("Τελικό loss:", loss_history[-1])
```

```python
#ΚΩΔΙΚΑΣ 5.4.2.4
from qiskit import QuantumCircuit

def encode_input(qc, x, n_qubits):
    """
    Κωδικοποίηση της εισόδου x στο κβαντικό κύκλωμα.
    Η ίδια τιμή x εφαρμόζεται σε όλα τα qubits μέσω της πύλης Ry(x).
    """
    for q in range(n_qubits):
        qc.ry(x, q)
    
    return qc
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.1
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

def quantum_model(x, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου
    encode_input(qc, x, n_qubits)

    # Παραμετροποιημένες πύλες
    for q in range(n_qubits):
        qc.ry(params[q], q)
        qc.rz(params[n_qubits + q], q)

    # Αλυσιδωτή εμπλοκή CNOT για n > 1
    for q in range(n_qubits - 1):
        qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο πρώτο qubit
    observable = SparsePauliOp("I" * (n_qubits - 1) + "Z")

    # Αναμενόμενη τιμή
    output = np.real(state.expectation_value(observable))

    return output
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.2
def loss_function(params, x_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model(x, params, n_qubits) for x in x_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.3
def train_random_search(x_train, y_train, n_qubits, 
                        iterations=1000, step_size=0.1):
    # Αρχικοποίηση παραμέτρων
    params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)

    best_params = params.copy()
    best_loss = loss_function(best_params, x_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):
        # Τυχαία μεταβολή των παραμέτρων
        candidate_params = best_params + step_size * (
            np.random.rand(2*n_qubits) - 0.5
        )

        candidate_loss = loss_function(
            candidate_params, x_train, y_train, n_qubits
        )

        # Κρατάμε τις νέες παραμέτρους μόνο αν μειώνουν το loss
        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.4
def parameter_shift_derivative(x, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2

    forward = quantum_model(x, shifted_plus, n_qubits)
    backward = quantum_model(x, shifted_minus, n_qubits)

    derivative = 0.5 * (forward - backward)

    return derivative


def compute_gradient(params, x_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    for k in range(len(params)):
        grad_k = 0

        for x, y in zip(x_data, y_data):
            y_pred = quantum_model(x, params, n_qubits)
            dy_dtheta = parameter_shift_derivative(
                x, params, n_qubits, k
            )

            grad_k += 2 * (y_pred - y) * dy_dtheta

        gradient[k] = grad_k / len(x_data)

    return gradient
```

```python
#ΚΩΔΙΚΑΣ 5.4.3.5
def train_gradient_descent(x_train, y_train, n_qubits, 
                           epochs=100, learning_rate=0.1):
    # Αρχικοποίηση παραμέτρων
    params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)

    loss_history = []

    for epoch in range(epochs):
        loss = loss_function(params, x_train, y_train, n_qubits)
        loss_history.append(loss)

        gradient = compute_gradient(params, x_train, y_train, n_qubits)

        # Ενημέρωση παραμέτρων
        params = params - learning_rate * gradient

    return params, loss_history
```

```python
# ΚΩΔΙΚΑΣ 5.4.3.6
# Εκπαίδευση όλων των QML μοντέλων για το πρόβλημα y = sin(2x)
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

# ------------------------------------------------------------
# 1. Δημιουργία δεδομένων
# ------------------------------------------------------------
seed = 42
np.random.seed(seed)
N = 40

# Είσοδος στο διάστημα [0, π]
x_data = np.linspace(0, np.pi, N)

# Συνάρτηση στόχος
y_data = np.sin(2 * x_data)

# Τυχαία αναδιάταξη δειγμάτων
indices = np.arange(N)
np.random.shuffle(indices)

# Train-test split
train_size = int(0.75 * N)

train_indices = indices[:train_size]
test_indices = indices[train_size:]

x_train = x_data[train_indices]
y_train = y_data[train_indices]

x_test = x_data[test_indices]
y_test = y_data[test_indices]

print("Πλήθος δειγμάτων εκπαίδευσης:", len(x_train))
print("Πλήθος δειγμάτων ελέγχου:", len(x_test))


# ------------------------------------------------------------
# 2. Κοινές πειραματικές ρυθμίσεις
# ------------------------------------------------------------

qubit_list = [1, 2, 4, 10]

learning_rate = 0.1
epochs = 100

random_iterations = 1000
random_step = 0.1


# ------------------------------------------------------------
# 3. Κωδικοποίηση εισόδου
# ------------------------------------------------------------

def encode_input(qc, x, n_qubits):
    """
    Κωδικοποίηση της εισόδου x στο κβαντικό κύκλωμα.
    Η ίδια τιμή x εφαρμόζεται σε όλα τα qubits μέσω της πύλης Ry(x).
    """
    for q in range(n_qubits):
        qc.ry(x, q)

    return qc


# ------------------------------------------------------------
# 4. Ορισμός QML μοντέλου
# ------------------------------------------------------------

def quantum_model(x, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση εισόδου
    encode_input(qc, x, n_qubits)

    # Παραμετροποιημένες πύλες Ry και Rz
    for q in range(n_qubits):
        qc.ry(params[q], q)
        qc.rz(params[n_qubits + q], q)

    # Αλυσιδωτή εμπλοκή CNOT για n > 1
    for q in range(n_qubits - 1):
        qc.cx(q, q + 1)

    # Υπολογισμός τελικής κατάστασης
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο πρώτο qubit q0
    # Στο Qiskit το δεξιότερο γράμμα αντιστοιχεί στο q0
    observable = SparsePauliOp("I" * (n_qubits - 1) + "Z")

    # Αναμενόμενη τιμή
    output = np.real(state.expectation_value(observable))

    return output


# ------------------------------------------------------------
# 5. Συνάρτηση κόστους
# ------------------------------------------------------------

def loss_function(params, x_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model(x, params, n_qubits) for x in x_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss


def predict_model(params, x_data, n_qubits):
    predictions = np.array([
        quantum_model(x, params, n_qubits) for x in x_data
    ])

    return predictions


# ------------------------------------------------------------
# 6. Εκπαίδευση με Random Search
# ------------------------------------------------------------

def train_random_search(x_train, y_train, n_qubits,
                        iterations=1000, step_size=0.1,
                        initial_params=None):

    if initial_params is None:
        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
    else:
        params = initial_params.copy()

    best_params = params.copy()
    best_loss = loss_function(best_params, x_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):
        candidate_params = best_params + step_size * (
            np.random.rand(2*n_qubits) - 0.5
        )

        candidate_loss = loss_function(
            candidate_params,
            x_train,
            y_train,
            n_qubits
        )

        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history


# ------------------------------------------------------------
# 7. Parameter-shift και Gradient Descent
# ------------------------------------------------------------

def parameter_shift_derivative(x, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2

    forward = quantum_model(x, shifted_plus, n_qubits)
    backward = quantum_model(x, shifted_minus, n_qubits)

    derivative = 0.5 * (forward - backward)

    return derivative


def compute_gradient(params, x_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    for x, y in zip(x_data, y_data):
        y_pred = quantum_model(x, params, n_qubits)

        for k in range(len(params)):
            dy_dtheta = parameter_shift_derivative(
                x,
                params,
                n_qubits,
                k
            )

            gradient[k] += 2 * (y_pred - y) * dy_dtheta

    gradient = gradient / len(x_data)

    return gradient


def train_gradient_descent(x_train, y_train, n_qubits,
                           epochs=100, learning_rate=0.1,
                           initial_params=None):

    if initial_params is None:
        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
    else:
        params = initial_params.copy()

    loss_history = []

    for epoch in range(epochs):
        loss = loss_function(params, x_train, y_train, n_qubits)
        loss_history.append(loss)

        gradient = compute_gradient(params, x_train, y_train, n_qubits)

        params = params - learning_rate * gradient

    return params, loss_history


# ------------------------------------------------------------
# 8. Εκπαίδευση όλων των μοντέλων
# ------------------------------------------------------------

results_regression = {}
summary_regression = []

for n_qubits in qubit_list:

    print("\n" + "=" * 60)
    print("Μοντέλο", n_qubits, "qubits")
    print("=" * 60)

    # Ίδιες αρχικές παράμετροι για Random Search και Gradient Descent
    initial_params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)

    # -----------------------------
    # Random Search
    # -----------------------------

    print("Εκπαίδευση με Random Search...")

    params_random, loss_random = train_random_search(
        x_train,
        y_train,
        n_qubits,
        iterations=random_iterations,
        step_size=random_step,
        initial_params=initial_params
    )

    train_loss_random = loss_function(
        params_random,
        x_train,
        y_train,
        n_qubits
    )

    test_loss_random = loss_function(
        params_random,
        x_test,
        y_test,
        n_qubits
    )

    results_regression[(n_qubits, "random_search")] = {
        "params": params_random,
        "loss_history": loss_random,
        "train_loss": train_loss_random,
        "test_loss": test_loss_random
    }

    summary_regression.append([
        n_qubits,
        "Random Search",
        train_loss_random,
        test_loss_random
    ])

    print("Train loss Random Search:", train_loss_random)
    print("Test loss Random Search:", test_loss_random)

    # -----------------------------
    # Parameter-shift + Gradient Descent
    # -----------------------------

    print("Εκπαίδευση με Parameter-shift + Gradient Descent...")

    params_gd, loss_gd = train_gradient_descent(
        x_train,
        y_train,
        n_qubits,
        epochs=epochs,
        learning_rate=learning_rate,
        initial_params=initial_params
    )

    train_loss_gd = loss_function(
        params_gd,
        x_train,
        y_train,
        n_qubits
    )

    test_loss_gd = loss_function(
        params_gd,
        x_test,
        y_test,
        n_qubits
    )

    results_regression[(n_qubits, "parameter_shift_gd")] = {
        "params": params_gd,
        "loss_history": loss_gd,
        "train_loss": train_loss_gd,
        "test_loss": test_loss_gd
    }

    summary_regression.append([
        n_qubits,
        "Parameter-shift + GD",
        train_loss_gd,
        test_loss_gd
    ])

    print("Train loss Parameter-shift + GD:", train_loss_gd)
    print("Test loss Parameter-shift + GD:", test_loss_gd)


# ------------------------------------------------------------
# 9. Συγκεντρωτική εμφάνιση αποτελεσμάτων
# ------------------------------------------------------------

print("\n\nΣΥΓΚΕΝΤΡΩΤΙΚΑ ΑΠΟΤΕΛΕΣΜΑΤΑ")
print("-" * 80)
print("Qubits | Μέθοδος Εκπαίδευσης | Train Loss | Test Loss")
print("-" * 80)

for row in summary_regression:
    n_qubits, method, train_loss, test_loss = row
    print(
        n_qubits, "|",
        method, "|",
        train_loss, "|",
        test_loss
    )
```

```python
#ΚΩΔΙΚΑΣ 5.5.3.1
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

def quantum_model_iris(x_vec, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση των δύο χαρακτηριστικών
    for q in range(n_qubits):
        qc.ry(x_vec[0], q)   # petal length
        qc.rz(x_vec[1], q)   # petal width

    # Παραμετροποιημένες πύλες
    for q in range(n_qubits):
        qc.ry(params[q], q)
        qc.rz(params[n_qubits + q], q)

    # Αλυσιδωτή εμπλοκή CNOT για n > 1
    for q in range(n_qubits - 1):
        qc.cx(q, q + 1)

    # Τελική κατάσταση
    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο πρώτο qubit q0
    observable = SparsePauliOp("I" * (n_qubits - 1) + "Z")

    output = np.real(state.expectation_value(observable))

    return output
```

```python
#ΚΩΔΙΚΑΣ 5.5.3.2
def iris_loss_function(params, X_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model_iris(x, params, n_qubits) for x in X_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss


def iris_predict_values(params, X_data, n_qubits):
    predictions = np.array([
        quantum_model_iris(x, params, n_qubits) for x in X_data
    ])

    return predictions


def iris_predict_classes(params, X_data, n_qubits):
    predictions = iris_predict_values(params, X_data, n_qubits)

    # Threshold στο 0
    predicted_classes = np.where(predictions >= 0, 1, 0)

    return predicted_classes


def iris_accuracy(params, X_data, y_class, n_qubits):
    predicted_classes = iris_predict_classes(params, X_data, n_qubits)

    accuracy = np.mean(predicted_classes == y_class)

    return accuracy
```

```python
#ΚΩΔΙΚΑΣ 5.5.3.3
def train_iris_random_search(X_train, y_train, n_qubits,
                             iterations=500, step_size=0.1,
                             initial_params=None):

    if initial_params is None:
        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
    else:
        params = initial_params.copy()

    best_params = params.copy()
    best_loss = iris_loss_function(best_params, X_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):
        candidate_params = best_params + step_size * (
            np.random.rand(2*n_qubits) - 0.5
        )

        candidate_loss = iris_loss_function(
            candidate_params,
            X_train,
            y_train,
            n_qubits
        )

        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history
```

```python
#ΚΩΔΙΚΑΣ 5.5.3.4
def iris_parameter_shift_derivative(x_vec, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2

    forward = quantum_model_iris(x_vec, shifted_plus, n_qubits)
    backward = quantum_model_iris(x_vec, shifted_minus, n_qubits)

    derivative = 0.5 * (forward - backward)

    return derivative

def iris_compute_gradient(params, X_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    for x_vec, y in zip(X_data, y_data):
        y_pred = quantum_model_iris(x_vec, params, n_qubits)

        for k in range(len(params)):
            dy_dtheta = iris_parameter_shift_derivative(
                x_vec,
                params,
                n_qubits,
                k
            )

            gradient[k] += 2 * (y_pred - y) * dy_dtheta

    gradient = gradient / len(X_data)

    return gradient

def train_iris_gradient_descent(X_train, y_train, n_qubits,
                                epochs=50, learning_rate=0.1,
                                initial_params=None):

    if initial_params is None:
        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
    else:
        params = initial_params.copy()

    loss_history = []

    for epoch in range(epochs):
        loss = iris_loss_function(params, X_train, y_train, n_qubits)
        loss_history.append(loss)

        gradient = iris_compute_gradient(params, X_train, y_train, n_qubits)

        params = params - learning_rate * gradient

    return params, loss_history
```

```python
#ΚΩΔΙΚΑΣ 5.6.3.1
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

def quantum_model_blobs(x_vec, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση των δύο χαρακτηριστικών
    for q in range(n_qubits):
        qc.ry(x_vec[0], q)
        qc.rz(x_vec[1], q)

    # Παραμετροποιημένες πύλες
    for q in range(n_qubits):
        qc.ry(params[q], q)
        qc.rz(params[n_qubits + q], q)

    # Αλυσιδωτή εμπλοκή CNOT
    for q in range(n_qubits - 1):
        qc.cx(q, q + 1)

    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο πρώτο qubit q0
    observable = SparsePauliOp("I" * (n_qubits - 1) + "Z")

    output = np.real(state.expectation_value(observable))

    return output
```

```python
#ΚΩΔΙΚΑΣ 5.6.3.2
def blobs_loss_function(params, X_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model_blobs(x, params, n_qubits) for x in X_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss


def blobs_predict_values(params, X_data, n_qubits):
    predictions = np.array([
        quantum_model_blobs(x, params, n_qubits) for x in X_data
    ])

    return predictions


def blobs_predict_classes(params, X_data, n_qubits):
    predictions = blobs_predict_values(params, X_data, n_qubits)

    predicted_classes = np.where(predictions >= 0, 1, 0)

    return predicted_classes


def blobs_accuracy(params, X_data, y_class, n_qubits):
    predicted_classes = blobs_predict_classes(params, X_data, n_qubits)

    accuracy = np.mean(predicted_classes == y_class)

    return accuracy
```

```python
# ΕΝΙΑΙΟΣ ΚΩΔΙΚΑΣ 5.6.3
# Synthetic Blobs - Εκπαίδευση QML μοντέλων 1, 2, 4 και 10 qubits
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp

# ------------------------------------------------------------
# 1. Δημιουργία και προεπεξεργασία δεδομένων
# ------------------------------------------------------------

seed = 42
np.random.seed(seed)

X_blobs, y_blobs = make_blobs(
    n_samples=120,
    centers=2,
    n_features=2,
    cluster_std=1.2,
    random_state=seed
)

# Κανονικοποίηση στο διάστημα [0, π]
scaler = MinMaxScaler(feature_range=(0, np.pi))
X_blobs_scaled = scaler.fit_transform(X_blobs)

# Μετατροπή labels από {0, 1} σε {-1, +1}
y_blobs_quantum = 2 * y_blobs - 1

# Train-test split
X_blobs_train, X_blobs_test, y_blobs_train_class, y_blobs_test_class, y_blobs_train, y_blobs_test = train_test_split(
    X_blobs_scaled,
    y_blobs,
    y_blobs_quantum,
    test_size=0.25,
    random_state=seed,
    stratify=y_blobs
)

print("Πλήθος συνολικών δειγμάτων:", len(X_blobs_scaled))
print("Πλήθος δειγμάτων εκπαίδευσης:", len(X_blobs_train))
print("Πλήθος δειγμάτων ελέγχου:", len(X_blobs_test))


# ------------------------------------------------------------
# 2. Πειραματικές ρυθμίσεις
# ------------------------------------------------------------

qubit_list = [1, 2, 4, 10]

blobs_random_iterations = 500
blobs_epochs = 50
learning_rate = 0.1
random_step = 0.1


# ------------------------------------------------------------
# 3. QML μοντέλο για Synthetic Blobs
# ------------------------------------------------------------

def quantum_model_blobs(x_vec, params, n_qubits):
    qc = QuantumCircuit(n_qubits)

    # Κωδικοποίηση των δύο χαρακτηριστικών
    for q in range(n_qubits):
        qc.ry(x_vec[0], q)
        qc.rz(x_vec[1], q)

    # Παραμετροποιημένες πύλες
    for q in range(n_qubits):
        qc.ry(params[q], q)
        qc.rz(params[n_qubits + q], q)

    # Αλυσιδωτή εμπλοκή CNOT
    for q in range(n_qubits - 1):
        qc.cx(q, q + 1)

    state = Statevector.from_instruction(qc)

    # Μέτρηση Pauli-Z στο πρώτο qubit q0
    observable = SparsePauliOp("I" * (n_qubits - 1) + "Z")

    output = np.real(state.expectation_value(observable))

    return output

# ------------------------------------------------------------
# 4. Loss, predictions και accuracy
# ------------------------------------------------------------

def blobs_loss_function(params, X_data, y_data, n_qubits):
    predictions = np.array([
        quantum_model_blobs(x, params, n_qubits) for x in X_data
    ])

    loss = np.mean((predictions - y_data) ** 2)

    return loss


def blobs_predict_values(params, X_data, n_qubits):
    predictions = np.array([
        quantum_model_blobs(x, params, n_qubits) for x in X_data
    ])

    return predictions


def blobs_predict_classes(params, X_data, n_qubits):
    predictions = blobs_predict_values(params, X_data, n_qubits)

    # Threshold στο 0
    predicted_classes = np.where(predictions >= 0, 1, 0)

    return predicted_classes


def blobs_accuracy(params, X_data, y_class, n_qubits):
    predicted_classes = blobs_predict_classes(params, X_data, n_qubits)

    accuracy = np.mean(predicted_classes == y_class)

    return accuracy

# ------------------------------------------------------------
# 5. Random Search
# ------------------------------------------------------------

def train_blobs_random_search(X_train, y_train, n_qubits,
                              iterations=500, step_size=0.1,
                              initial_params=None):

    if initial_params is None:
        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
    else:
        params = initial_params.copy()

    best_params = params.copy()
    best_loss = blobs_loss_function(best_params, X_train, y_train, n_qubits)

    loss_history = []

    for i in range(iterations):
        candidate_params = best_params + step_size * (
            np.random.rand(2*n_qubits) - 0.5
        )

        candidate_loss = blobs_loss_function(
            candidate_params,
            X_train,
            y_train,
            n_qubits
        )

        if candidate_loss < best_loss:
            best_loss = candidate_loss
            best_params = candidate_params.copy()

        loss_history.append(best_loss)

    return best_params, loss_history


# ------------------------------------------------------------
# 6. Parameter-shift + Gradient Descent
# ------------------------------------------------------------

def blobs_parameter_shift_derivative(x_vec, params, n_qubits, param_index):
    shifted_plus = params.copy()
    shifted_minus = params.copy()

    shifted_plus[param_index] += np.pi / 2
    shifted_minus[param_index] -= np.pi / 2

    forward = quantum_model_blobs(x_vec, shifted_plus, n_qubits)
    backward = quantum_model_blobs(x_vec, shifted_minus, n_qubits)

    derivative = 0.5 * (forward - backward)

    return derivative


def blobs_compute_gradient(params, X_data, y_data, n_qubits):
    gradient = np.zeros_like(params)

    for x_vec, y in zip(X_data, y_data):
        y_pred = quantum_model_blobs(x_vec, params, n_qubits)

        for k in range(len(params)):
            dy_dtheta = blobs_parameter_shift_derivative(
                x_vec,
                params,
                n_qubits,
                k
            )

            gradient[k] += 2 * (y_pred - y) * dy_dtheta

    gradient = gradient / len(X_data)

    return gradient


def train_blobs_gradient_descent(X_train, y_train, n_qubits,
                                 epochs=50, learning_rate=0.1,
                                 initial_params=None):

    if initial_params is None:
        params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)
    else:
        params = initial_params.copy()

    loss_history = []

    for epoch in range(epochs):
        loss = blobs_loss_function(params, X_train, y_train, n_qubits)
        loss_history.append(loss)

        gradient = blobs_compute_gradient(params, X_train, y_train, n_qubits)

        params = params - learning_rate * gradient

    return params, loss_history


# ------------------------------------------------------------
# 7. Εκπαίδευση όλων των μοντέλων
# ------------------------------------------------------------

results_blobs = {}
summary_blobs = []

for n_qubits in qubit_list:

    print("\n" + "=" * 60)
    print("Synthetic Blobs model:", n_qubits, "qubits")
    print("=" * 60)

    # Ίδιες αρχικές παράμετροι για δίκαιη σύγκριση των δύο μεθόδων
    initial_params = np.random.uniform(0, 2*np.pi, size=2*n_qubits)

    # --------------------------------------------------
    # Random Search
    # --------------------------------------------------

    print("Εκπαίδευση με Random Search...")

    params_random, loss_random = train_blobs_random_search(
        X_blobs_train,
        y_blobs_train,
        n_qubits,
        iterations=blobs_random_iterations,
        step_size=random_step,
        initial_params=initial_params
    )

    train_loss_random = blobs_loss_function(
        params_random,
        X_blobs_train,
        y_blobs_train,
        n_qubits
    )

    test_loss_random = blobs_loss_function(
        params_random,
        X_blobs_test,
        y_blobs_test,
        n_qubits
    )

    train_acc_random = blobs_accuracy(
        params_random,
        X_blobs_train,
        y_blobs_train_class,
        n_qubits
    )

    test_acc_random = blobs_accuracy(
        params_random,
        X_blobs_test,
        y_blobs_test_class,
        n_qubits
    )

    results_blobs[(n_qubits, "random_search")] = {
        "params": params_random,
        "loss_history": loss_random,
        "train_loss": train_loss_random,
        "test_loss": test_loss_random,
        "train_accuracy": train_acc_random,
        "test_accuracy": test_acc_random
    }

    summary_blobs.append([
        n_qubits,
        "Random Search",
        train_loss_random,
        test_loss_random,
        train_acc_random,
        test_acc_random
    ])

    print("Train loss Random Search:", train_loss_random)
    print("Test loss Random Search:", test_loss_random)
    print("Train accuracy Random Search:", train_acc_random)
    print("Test accuracy Random Search:", test_acc_random)

    # --------------------------------------------------
    # Parameter-shift + Gradient Descent
    # --------------------------------------------------

    print("Εκπαίδευση με Parameter-shift + Gradient Descent...")

    params_gd, loss_gd = train_blobs_gradient_descent(
        X_blobs_train,
        y_blobs_train,
        n_qubits,
        epochs=blobs_epochs,
        learning_rate=learning_rate,
        initial_params=initial_params
    )

    train_loss_gd = blobs_loss_function(
        params_gd,
        X_blobs_train,
        y_blobs_train,
        n_qubits
    )

    test_loss_gd = blobs_loss_function(
        params_gd,
        X_blobs_test,
        y_blobs_test,
        n_qubits
    )

    train_acc_gd = blobs_accuracy(
        params_gd,
        X_blobs_train,
        y_blobs_train_class,
        n_qubits
    )

    test_acc_gd = blobs_accuracy(
        params_gd,
        X_blobs_test,
        y_blobs_test_class,
        n_qubits
    )

    results_blobs[(n_qubits, "parameter_shift_gd")] = {
        "params": params_gd,
        "loss_history": loss_gd,
        "train_loss": train_loss_gd,
        "test_loss": test_loss_gd,
        "train_accuracy": train_acc_gd,
        "test_accuracy": test_acc_gd
    }

    summary_blobs.append([
        n_qubits,
        "Parameter-shift + GD",
        train_loss_gd,
        test_loss_gd,
        train_acc_gd,
        test_acc_gd
    ])

    print("Train loss Parameter-shift + GD:", train_loss_gd)
    print("Test loss Parameter-shift + GD:", test_loss_gd)
    print("Train accuracy Parameter-shift + GD:", train_acc_gd)
    print("Test accuracy Parameter-shift + GD:", test_acc_gd)


# ------------------------------------------------------------
# 8. Συγκεντρωτικά αποτελέσματα
# ------------------------------------------------------------

print("\n\nΣΥΓΚΕΝΤΡΩΤΙΚΑ ΑΠΟΤΕΛΕΣΜΑΤΑ SYNTHETIC BLOBS")
print("-" * 120)
print("Qubits | Μέθοδος | Train Loss | Test Loss | Train Accuracy | Test Accuracy")
print("-" * 120)

for row in summary_blobs:
    n_qubits, method, train_loss, test_loss, train_acc, test_acc = row

    print(
        n_qubits, "|",
        method, "|",
        train_loss, "|",
        test_loss, "|",
        train_acc, "|",
        test_acc
    )
```
